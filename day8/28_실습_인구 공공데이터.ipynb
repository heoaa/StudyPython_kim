```python
import pandas as pd
```

#### 데이터 불러오기

```python
# thousands : 숫자 천단위 자리 콤마 제외하고 불러옴
df = pd.read_csv('people_202101.csv', encoding = 'cp949', thousands = ',')
df.head(2)
```

#### 행정구역 정리

```python
df['행정구역_시'] = df['행정구역'].apply(lambda x: x.split(' ')[0])
df['행정구역_구'] = df['행정구역'].apply(lambda x: x.split(' ')[1])
df['행정구역_동'] = df['행정구역'].apply(lambda x: x.split(' ')[2])
# df['행정구역_코드'] = df['행정구역'].apply(lambda x: x.split('(').rstrip(')')[3])
```

```python
df2 = df[df['행정구역'] == '서울특별시 관악구 신림동(1162069500)']
df2
```

#### 불필요 컬럼 삭제

```python
df2 = df2.drop(['행정구역','2021년01월_계_총인구수','2021년01월_계_연령구간인구수','행정구역_시','행정구역_구'], axis = 1)
df2
```

```python
df3 = df2.melt()
```

```python
import matplotlib.pyplot as plt

plt.plot(df3)
plt.show()
```
