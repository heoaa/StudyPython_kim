* 브로드캐스팅(Broadcasting)

```python
print(scientists*2)   # 문자열의 경우 반복연산
```

- 열의 자료형 바꾸기

```python
# 열의 자료형 확인
print(scientists['Born'].dtype)
print(scientists['Died'].dtype)
```

```python
# 'Born' 날짜형식 문자열을 datetime 자료형으로 변경
born_datetime = pd.to_datetime(scientists['Born'], format = '%Y-%m-%d')
print(born_datetime)
```

```python
# 'Died' 날짜형식 문자열을 datetime 자료형으로 변경
died_datetime = pd.to_datetime(scientists['Died'], format = '%Y-%m-%d')
print(died_datetime)
```

* 새로운 열 추가

```python
# 위의 datetime으로 바뀐 2개의 자료를 새로운 열로 추가
scientists['Born_dt'] = born_datetime
scientists['Died_dt'] = died_datetime
scientists
```

```python
# 자료형 확인
scientists.info()
```

```python
# 필요한 컬럼으로 df 재생성
scientists2 = scientists.loc[:, ['Name', 'Born_dt', 'Died_dt', 'Age', 'Occupation']]
scientists2
```

```python
# 과학자의 lifetime(age_days_dt) 계산 ( died_dt - born_dt)
scientists2['age_days_dt'] = ( scientists2['Died_dt'] - scientists2['Born_dt'] )
scientists2
```

* 데이터 삭제 : drop(['인덱스 또는 열'][ , axis=n]

```python
# 인덱스(행) 삭제
scientists.drop(0)
```

```python
scientists.drop([0,3])
```

```python
# 열 삭제
scientists.drop('Name', axis = 1)      # 0 : 행 삭제(수평),  1 : 열 삭제(수직)
```

```python
# 여러개의 열 삭제
scientists.drop(['Age', 'Occupation'], axis = 1)
```

* 행 또는 열 삭제 후 데이터 바로 적용 : inplace = True

```python
# 삭제 후 데이터 적용
scientists3 = scientists2.copy()    # scientists2가 바뀌어도(양방향 영향) scientists3이 변하지 않음 / (카피)
scientists3
```
