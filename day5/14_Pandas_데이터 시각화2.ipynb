## tips dataset 활용

```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.info()
```

### 히스토그램
* 일변량 그래프 : 하나의 변수를 사용한 그래프

```python
# 그래프 틀( 캔버스 크기, 해상도 지정 가능 )
fig = plt.figure()   #fig = plt.figure(figsize = (20,10), dpi = 600)
axes1 = fig.add_subplot(1,1,1)

# 그래프 작성
axes1.hist(tips['total_bill'], bins = 10, color = 'gray')    # bins : x축 간격 조정
axes1.set_title('Histogram of Total Bill')    # 제목
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

### 산점도 그래프 
* ! 이변량 그래프 : 변수 2개를 사용한 그래프

```python
# total_bill에 따른 tip 데이터 분포
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)

axes1.scatter(tips['total_bill'], tips['tip'], c='magenta')    # 변수는 x, y 순서대로 기입
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

### 박스 그래프

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
               tips[tips['sex'] == 'Male']['tip']],
              labels = ['Female','Male'])
axes1.set_title('Boxplot of Tips by Gender')
axes1.set_xlabel('Gender')
axes1.set_ylabel('Tip')
```

### 막대 그래프

```python
# bar 그래프
bar_plot = plt.figure()
axes1 = bar_plot.add_subplot(1,1,1)

axes1.bar(tips['sex'], tips['total_bill'])
```

### 이변량 그래프
* 산점도 표현

```python
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill',
                 y = 'tip',
                 data=tips)
ax.set_title('Scaotterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = sns.regplot(x = 'total_bill',
                 y = 'tip',
                 data = tips,
                 fit_reg = False)    # 회귀선 제거
```

* 산점도와 히스토그램 한번에 표현

```python
joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.05)   # y : 타이틀과 그래프 간격
```

```python
joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')   # 산점도 종류 변경(육각그래프)
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.05)   # y : 타이틀과 그래프 간격
```

### 다변량 그래프 : 3개 이상의 변수를 사용한 그래프
* 산점도 표현

```python
# 성별에 따라 마커의 색상을 변경하기 위한 함수 작성
def recode_gender(gender) :
    if gender == 'Female' :
        return 'red'
    else :
        return 'blue'
```

```python
# apply 함수를 사용하여 데이터를 변경한 후 color 열 추가
tips['color'] = tips['sex'].apply(recode_gender)
tips.head()
```

```python
# 인원(size)은 마커 크기로, 성별은 마커 색으로 표현
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)

axes1.scatter(x = tips['total_bill'],
              y = tips['tip'],
              s = tips['size']*30,
              c = tips['color'],
              alpha = 0.5)             # 그래프 투명도
axes1.set_title('Total Bill vs Tip Colored by Gender and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
scatter = sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', fit_reg = False)
```
