### 누락값
* 누락값 확인(NaN, NAN, nan)

```python
import numpy as np
```

```python
# True, False, 0 ,'' 과 달리 값 자체가 없는 값
print(np.NaN == True)
print(np.NaN == False)
print(np.NaN == 0)
print(np.NaN == '')
```

```python
# 값 자체가 없으므로 자기 자신과 비교해도 False 출력
print(np.NaN == np.NaN)
print(np.NaN == np.NaN)
print(np.NaN == np.NaN)
print(np.NaN == np.NaN)
```

* 누락값 확인 메소드 : isnull ,  notnull

```python
import pandas as pd
```

```python
# null인지 아닌지 판단 (nan값인 경우 True)
print(pd.isnull(np.NaN))
print(pd.isnull(np.NAN))
print(pd.isnull(np.nan))
```

```python
# (nan값인 경우 False)
print(pd.notnull(np.NaN))
print(pd.notnull(42))
print(pd.notnull('missing'))
```

#### 누락값 개수 확인 
* count()

```python
ebola = pd.read_csv('./country_timeseries.csv')
ebola.head()
```

```python
ebola.info()
```

```python
ebola.count()
```

```python
# shape[0]에 전체 행개수 - 값의 개수 = 누락값 개수 (브로드캐스팅)
num_rows = ebola.shape[0]
num_missing = num_rows - ebola.count()
print(num_missing)
```

* count_nonzero() : 배열에서 0(False)이 아닌 값의 개수 확인

```python
import numpy as np

print(np.count_nonzero(ebola.isnull()))
print(np.count_nonzero(ebola['Cases_Guinea'].isnull()))
```

```python
ebola.isnull()
```

* value_counts() : 지정한 열의 빈도 확인

```python
print(ebola['Cases_Guinea'].value_counts(), '\n')
print(ebola['Cases_Guinea'].value_counts(dropna=True))
```

#### 누락값 변경하기
* fillna() : 누락값을 지정한 값으로 변경

```python
ebola.head()
```

```python
ebola.fillna(0).head()
```

```python
print(ebola['Cases_Guinea'].sum())
print(ebola['Cases_Guinea'].fillna(0).sum())
```

```python
print(ebola['Cases_Guinea'].mean())
print(ebola['Cases_Guinea'].fillna(0).mean())
```

* fillna(method = ' ') 

```python
# ffill
print(ebola.iloc[:5,0:5])
print(ebola.fillna(method = 'ffill').iloc[:5,0:5])
```

```python
# bfill
print(ebola.iloc[:5,0:5])
print(ebola.fillna(method = 'bfill').iloc[:5,0:5])
```

* interpolate() : 누락값 양쪽에 있는 값을 이용하여 중간값을 구한 다음 변경

```python
ebola.iloc[:, 0:5].head()
```

```python
ebola.interpolate().iloc[:, 0:5]
```

```python
# 상황에 따라 너무 많은 데이터가 삭제될 수 있으므로 신중해야 함
print(ebola.shape)
ebola_dropna = ebola.dropna()
print(ebola_dropna.shape)
print(ebola_dropna)
```
